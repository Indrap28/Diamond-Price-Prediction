```python
from sklearn.preprocessing import LabelEncoder
```

```python
data_final=data.apply(LabelEncoder().fit_transform)
```

```python
data_final.head()
```

```python
X = data_final.drop(['price'], axis=1)
y = data_final['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
Norm=MinMaxScaler()
```

```python
X_train=Norm.fit_transform(X_train)
X_test=Norm.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
# Model Building
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)

```

```python

y_pred=lr.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
r2_lr=r2_score(y_test,y_pred)
mse_lr=mean_squared_error(y_test,y_pred)
```

```python
print(f'r2_Score is {r2_lr}')
print(f'Mean Squared error is {mse_lr}')
```

```python
#K Neighbors Regressor
from sklearn.neighbors import KNeighborsRegressor
kn=KNeighborsRegressor(n_neighbors=2)
kn.fit(X_train,y_train)
```

```python
y_pred=kn.predict(X_test)
```

```python
r2_kn=r2_score(y_test,y_pred)
mse_kn=mean_squared_error(y_test,y_pred)
```

```python
print(f'r2_Score is {r2_kn}')
print(f'Mean Squared error is {mse_kn}')
```

```python
# Decison Tree Regressor
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor(random_state = 0)
dt.fit(X_train,y_train)
```

```python
y_pred=dt.predict(X_test)

```

```python
pred=dt.predict([[ 2, 1, 3, 83, 30,19, 25,23,5]])
```

```python
pred[0]
```

```python

r2_dt=r2_score(y_test,y_pred)
mse_dt=mean_squared_error(y_test,y_pred)
```

```python
print(f'r2_Score is {r2_dt}')
print(f'Mean Squared error is {mse_dt}')
 
```

```python
#Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(max_depth=2, random_state=0)
rf.fit(X_train,y_train)


```

```python
y_pred=rf.predict(X_test)
```

```python
r2_rf=r2_score(y_test,y_pred)
mse_rf=mean_squared_error(y_test,y_pred)
```

```python
print(f'r2_Score is {r2_rf}')
print(f'Mean Squared error is {mse_rf}')
```

```python
# Gradient Boosting Regressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbr = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1,max_depth=1, random_state=0, loss='ls',verbose = 1)
gbr.fit(X_train,y_train)


```

```python

y_pred=gbr.predict(X_test)
```

```python
r2_gbr=r2_score(y_test,y_pred)
mse_gbr=mean_squared_error(y_test,y_pred)
```

```python
print(f'r2_Score is {r2_gbr}')
print(f'Mean Squared error is {mse_gbr}')
```

```python
# Adaboost Regression
from sklearn.ensemble import AdaBoostRegressor
```

```python
ada_reg = AdaBoostRegressor(random_state=0, n_estimators=100)
ada_reg.fit(X_train,y_train)
```

```python
y_pred=ada_reg.predict(X_test)
```

```python
r2_ada=r2_score(y_test,y_pred)
mse_ada=mean_squared_error(y_test,y_pred)
```

```python
r2_ada_regr=r2_score(y_test,y_pred)
mse_ada_regr=mean_squared_error(y_test,y_pred)

print(f'r2_Score is {r2_ada_regr}')
print(f'Mean Squared error is {mse_ada_regr}')
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(X_train,y_train)
```

```python
y_pred=lasso.predict(X_test)
```

```python
r2_lasso=r2_score(y_test,y_pred)
mse_lasso=mean_squared_error(y_test,y_pred)
```

```python
print(f'r2_Score is {r2_lasso}')
print(f'Mean Squared error is {mse_lasso}')
```
